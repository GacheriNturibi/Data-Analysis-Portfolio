###•	Identify most frequent accident type; Plot static/interactive graph showing this

In this case we are considering the VICTIM_TYPE column. Let's explore it further

```python
HDX_clean['VICTIM_TYPE'].unique()
```

There are many varied categories. With the many spelling errors, combinations etc, we need to use regular expressions (Regex). In my case I used an online regex generator to create regular expressions for my predefined categories:


1.   Rider - this is a motor cyclist
1.   Passenger
1.   Driver
1.   Pedestrian
2.   Cyclist - refers to a pedal cyclist (bicycle)
2.   Rider & Passenger
2.   Driver & Passenger
2.   Pedestrian & Rider

**NB:**
There are many assumptions and(or) generalizations made when coming up with the categories, hence this data cannot be fully relied on. A more manual approach for cleaning the column would be preferred. For visualization purposes, we shall use this as is.


```python
import re

def clean_category(category):
    # first let's convert category to lowercase
    category = category.lower()

    # we then use regex to extract relevant keywords
    rider = re.search(r'^(rider|re|rider$|m/cyclist|m/cyclit|m/ cyclist|riders)$', category)
    driver = re.search(r'^(driver)$', category)
    pedestrian = re.search(r'^(pedstiran|pedestrians|pedestrian|pedestrain)$', category)
    cyclist = re.search(r'^(pedal cyclist|p/cyclist)$', category)
    passenger = re.search(r'^(passenger|passanger|p/pasenger|p/paseenger|p/pasengers|p/pasenger|passenger|p/passenger|2 passengers & p/passenger)$', category)
    rider_passenger = re.search(r'^(rider & passenger|rider and passenger|rider p/pasanger|rider & p/pasenger||rider & p/passenger|rider and p/passenger|rider p/passanger|m/cyclist and p/pasenger|m/cyclist & p/pasenger|1 rider & 1p/pasenger|1 rider & 1p/passenger)$', category)
    driver_passenger = re.search(r'^(driver & passenger|driver and passenger|driver & 2 passengers|2 drivers & passenger|driver passenger|driver 2 passangers|7 passengers 1 driver|2 driver & 2 passenger|driver & passenger|1 driver & 4 passengers|1 driver & 2 passengers)$', category)
    pedestrian_rider = re.search(r'^(pedestrian & rider)$', category)

    # categorize based on which keyword was found
    if rider:
        return 'rider'
    elif passenger:
        return 'passenger'
    elif driver:
        return 'driver'
    elif pedestrian:
        return 'pedestrian'
    elif cyclist:
        return 'cyclist'
    elif rider_passenger:
        return 'rider & passenger'
    elif driver_passenger:
        return 'driver & passenger'
    elif pedestrian_rider:
        return 'pedestrian & rider'
    else:
        return 'other'
```

```python
# Apply the function while to a new column
HDX_clean['CLEAN_VICTIM_TYPE'] = HDX_clean['VICTIM_TYPE'].apply(clean_category)
HDX_clean
```

We now have a column, CLEAN_VICTIM_TYPE, with the 9 categories. Let's use it to plot the most common accident type.

```python
# Group the dataframe by the new column
type_counts = HDX_clean.groupby('CLEAN_VICTIM_TYPE').size().reset_index(name='Number of Accidents')

# Bar Plot
fig = px.bar(type_counts, x='CLEAN_VICTIM_TYPE', y='Number of Accidents', color='CLEAN_VICTIM_TYPE',
             color_discrete_sequence=px.colors.qualitative.Plotly) # There are different color paletters for the plotly library.

# Update the layout
fig.update_layout(
   title="Accident Types",
   xaxis_title="Victim Type",
   legend_title="Victim Type",
   )
# Show the plot
fig.show()
```

From this plot, we notice that the most frequent accidents are those that involve **pedestrians with 282 accidents**. They are then followed by passenger accidents (138) and then motorcycle accidents (125).

###•	Summary of affected gender by accident type; Plot static/interactive graph showing this

```python
# Let's explore the gender column
HDX_clean['GENDER'].unique()
```

```python
# The genders are mixed up. Let's clean it up to have 4 categories
# Male, Female, Both, Unknown
def clean_gender(gender):

    # we use regex to extract relevant keywords
    male = re.search(r'^(M|M/A|M/J)$', gender)
    female = re.search(r'^(F|F/J)$', gender)
    both = re.search(r'^(M & F/J|M&F|M & F|M AND F|M/F|2M & F|M&f)$', gender)

    # categorize based on which keyword was found
    if male:
        return 'Male'
    elif female:
        return 'Female'
    elif both:
        return 'Both'
    else:
        return 'Unknown'
```

```python
# Apply the function to a new column
HDX_clean['CLEAN_GENDER'] = HDX_clean['GENDER'].apply(clean_gender)
HDX_clean
```

```python
# Group the dataframe by victim type and gender, and count the number of accidents for each group
type_gender_counts = HDX_clean.groupby(['CLEAN_VICTIM_TYPE', 'CLEAN_GENDER']).size().reset_index(name='Number of Accidents')

# Create a bar plot using plotly
fig = px.bar(type_gender_counts, x='CLEAN_VICTIM_TYPE', y='Number of Accidents', color='CLEAN_GENDER',
             color_discrete_sequence=['blue', 'pink', 'green', 'grey'], # Set colors for male and female and the rest
             barmode='group') # Group bars by victim type

# Update the layout
fig.update_layout(
   title="Accident Types by Gender",
   xaxis_title="Victim Type",
   legend_title="Victim By Gender",
   )

# Show the plot
fig.show()
```

We notice that in all accident types, there are **more male victims** than there are females and both genders. **All pedal cyclists are male**, while **all but one motorcyclists are male**. This is an interesting statistic!
